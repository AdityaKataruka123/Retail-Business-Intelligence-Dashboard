```python
df_clean = df.copy()

print("✅ Working copy created successfully!")
```

```python
df_clean.columns = (
    df_clean.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df_clean.columns
```

```python
df_clean.isnull().sum()
```

```python
duplicates_before = df_clean.duplicated().sum()

print("Duplicate Rows Before:", duplicates_before)

df_clean.drop_duplicates(inplace=True)

duplicates_after = df_clean.duplicated().sum()

print("Duplicate Rows After :", duplicates_after)
```

```python
df_clean["order_date"] = pd.to_datetime(df_clean["order_date"])
df_clean["ship_date"] = pd.to_datetime(df_clean["ship_date"])

print(df_clean.dtypes)
```

```python
df_clean["order_year"] = df_clean["order_date"].dt.year
df_clean["order_month"] = df_clean["order_date"].dt.month_name()
df_clean["order_quarter"] = df_clean["order_date"].dt.quarter
df_clean["order_day"] = df_clean["order_date"].dt.day_name()

df_clean.head()
```

```python
df_clean.to_csv("cleaned_superstore (1).csv", index=False)
```

```python
plt.style.use("ggplot")

plt.rcParams["figure.figsize"] = (12,6)
plt.rcParams["font.size"] = 12

print("Visualization settings applied successfully!")
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df_clean["sales"], bins=30, kde=True)

plt.title("Distribution of Sales", fontsize=18)
plt.xlabel("Sales")
plt.ylabel("Frequency")

plt.show()
```

## 📌 Business Insight

Most retail transactions involve relatively low sales amounts, while a small number of transactions contribute significantly higher sales.

This indicates that the sales distribution is positively skewed, which is common in retail businesses.

```python
plt.figure(figsize=(10,6))

sns.histplot(df_clean["profit"], bins=30, kde=True)

plt.title("Distribution of Profit", fontsize=18)
plt.xlabel("Profit")
plt.ylabel("Frequency")

plt.show()
```

## 📌 Business Insight

The profit distribution contains both positive and negative values, indicating that some transactions resulted in financial losses.

Further analysis is required to identify the categories and regions contributing to negative profits.

```python
category_sales = (
    df_clean.groupby("category")["sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=category_sales.index,
    y=category_sales.values
)

plt.title("Total Sales by Category", fontsize=18)
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

## 📌 Business Insight

Comparing total sales across product categories helps identify the company's primary revenue drivers.

Management can use this information to prioritize inventory planning and marketing investments.

```python
category_profit = (
    df_clean.groupby("category")["profit"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=category_profit.index,
    y=category_profit.values
)

plt.title("Total Profit by Category", fontsize=18)
plt.xlabel("Category")
plt.ylabel("Total Profit")

plt.show()
```

## 📌 Business Insight

The analysis of profit by category reveals which product categories contribute the most to the company's overall profitability. While high sales are important, they do not always translate into high profits.

Categories with lower profits despite strong sales may indicate issues such as excessive discounts, high operational costs, or lower profit margins. Management should focus on maximizing profitability by optimizing pricing strategies and improving the performance of less profitable categories.

```python
monthly_sales = (
    df_clean.groupby("order_month")["sales"]
    .sum()
    .reindex([
        "January", "February", "March", "April", "May", "June",
        "July", "August", "September", "October", "November", "December"
    ])
)

plt.figure(figsize=(12,6))

plt.plot(
    monthly_sales.index,
    monthly_sales.values,
    marker="o",
    linewidth=3
)

plt.title("Monthly Sales Trend", fontsize=18, weight="bold")
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

## 📌 Business Insight

The monthly sales trend helps identify seasonal patterns in customer purchasing behavior.

Months with higher sales indicate peak business periods, while months with lower sales may require targeted marketing campaigns or promotional offers.

```python
region_sales = (
    df_clean.groupby("region")["sales"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=region_sales.index,
    y=region_sales.values
)

plt.title("Sales by Region", fontsize=18, weight="bold")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

## 📌 Business Insight

Regional sales analysis highlights the geographical contribution to overall revenue.

High-performing regions represent strong markets, while lower-performing regions may require additional sales strategies and market expansion efforts.

```python
region_profit = (
    df_clean.groupby("region")["profit"]
    .sum()
    .sort_values(ascending=False)
)

plt.figure(figsize=(8,5))

sns.barplot(
    x=region_profit.index,
    y=region_profit.values
)

plt.title("Profit by Region", fontsize=18, weight="bold")
plt.xlabel("Region")
plt.ylabel("Total Profit")

plt.show()
```

## 📌 Business Insight

Profitability varies across different regions due to differences in customer demand, operational costs, and discount strategies.

Regions generating high sales but lower profits may require a review of pricing and operational efficiency.

# 📊 Executive Business KPI Summary

This section provides a high-level overview of the business using key performance indicators (KPIs). These metrics help management quickly evaluate overall business performance before exploring detailed visualizations.

```python
total_sales = df_clean["sales"].sum()
total_profit = df_clean["profit"].sum()
total_orders = df_clean["order_id"].nunique()
total_customers = df_clean["customer_id"].nunique()
average_discount = df_clean["discount"].mean()
average_sales = df_clean["sales"].mean()
profit_margin = (total_profit / total_sales) * 100

print("="*50)
print("📊 EXECUTIVE BUSINESS SUMMARY")
print("="*50)

print(f"💰 Total Sales        : ${total_sales:,.2f}")
print(f"📈 Total Profit       : ${total_profit:,.2f}")
print(f"📦 Total Orders       : {total_orders}")
print(f"👥 Total Customers    : {total_customers}")
print(f"🛒 Average Sale       : ${average_sales:.2f}")
print(f"💸 Average Discount   : {average_discount:.2%}")
print(f"📊 Profit Margin      : {profit_margin:.2f}%")
```

## 📌 Business Insight

The executive KPI summary provides a quick overview of the company's overall performance.

These metrics serve as the foundation for strategic decision-making and will later be displayed as KPI cards in the Power BI dashboard.

```python
monthly_summary = (
    df_clean.groupby("order_month")[["sales", "profit"]]
    .sum()
    .reindex([
        "January","February","March","April","May","June",
        "July","August","September","October","November","December"
    ])
)

monthly_summary.plot(
    figsize=(12,6),
    marker="o",
    linewidth=3
)

plt.title("Monthly Sales & Profit Trend", fontsize=18, weight="bold")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.xticks(rotation=45)

plt.show()
```

## 📌 Business Insight

Comparing monthly sales with profit helps determine whether periods of high sales also generate strong profitability.

This analysis can reveal months where aggressive discounts or increased costs reduced overall profit despite strong sales.

```python
loss_products = (
    df_clean.groupby("product_name")["profit"]
    .sum()
    .sort_values()
    .head(10)
)

plt.figure(figsize=(12,6))

sns.barplot(
    x=loss_products.values,
    y=loss_products.index
)

plt.title("Top 10 Loss-Making Products", fontsize=18, weight="bold")
plt.xlabel("Profit")
plt.ylabel("Product")

plt.show()
```

## 📌 Business Insight

Some products consistently generate financial losses despite being sold.

These products should be reviewed for pricing strategy, supplier costs, or discount policies.
